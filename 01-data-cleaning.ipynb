```python
import numpy as np
import pandas as pd
```

```python
ds = pd.read_csv("data.csv")
```

```python
ds.info()
```

```python
ds["year"].unique()
```

```python
ds['year'] = pd.to_numeric(ds['year'], errors = 'coerce')# first make it numeric
ds = ds.dropna(subset=['year'])#drop nan 
ds['year'] = ds['year'].astype(int) # convert datatype to int(integer) because year cannot be in float
```

```python
ds["Price"] = ds["Price"].str.replace(",","")
ds["Price"].unique()
```

```python
ds["Price"] = pd.to_numeric(ds["Price"] , errors = 'coerce')
```

```python
ds["Price"].unique()
```

```python
ds.dropna(inplace=True)
```

```python
ds["kms_driven"] = ds["kms_driven"].str.replace(",","")
ds["kms_driven"] = ds["kms_driven"].str.replace(" kms","")
ds["kms_driven"].unique()
```

```python
ds["kms_driven"] = pd.to_numeric(ds["kms_driven"] , errors = 'coerce')
ds["kms_driven"]
```

```python
#to remove unreasonable range or new cars
ds = ds[ds["kms_driven"] >= 10000]
```

```python
# remove extremely overused cars or errors
ds = ds[ds["kms_driven"] <= 200000]
```

```python
ds.info()
```

```python
ds["fuel_type"].unique()
```

```python
ds["company"].unique()
```

```python
ds["name"].unique()
```

```python
for i in ds.index:
    company = ds.loc[i,"company"]
    name = ds.loc[i,"name"]
    name  = name.replace(company,"").strip() 
    ds.loc[i, "name"] = name # change in dataset
    print(name)
```

```python
ds.duplicated().unique()
```

```python
ds.drop_duplicates(inplace = True)
```

```python
ds.info()
```

```python
#used to reset the index back to default integer index
ds.reset_index(inplace=True)
```

```python
ds.to_csv("cleaned_car_data.csv")
```
